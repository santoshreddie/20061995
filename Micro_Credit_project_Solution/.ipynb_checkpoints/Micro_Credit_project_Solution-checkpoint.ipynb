# Correlation

```python
#There are many columns in the dataset so lets create function such that we are setting a threshold, 
#so that we will find the high correlated columns and lets drop them
def correlation(dataset, threshold):
    col_corr = set() #set of all correlated values
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j] > threshold):
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

# Outlier detection and Transformation

```python
#In machine learning outlier detection is most important for building an accutate model to get good results
#lets create a function that deals with checking outliers
def plotingvariables(df, column):
    plt.figure(figsize=(16,4))
    plt.subplot(1,2,1)
    plt.hist(df[column], alpha = 0.5)
    plt.title(column)

    plt.subplot(1,2,2)
    plt.boxplot(df[column])
    plt.show()
    
    
#creating the boxplots for the data
for i in data:
    plotingvariables(data, i)
```

```python
#here ia mcreating function where i am getting uper and lower bound of the data
for col in data:
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    
    iqr = q3-q1
    
    lower_bridge = q3 + (1.5*iqr)
    
    upper_bridge = q1 - (1.5*iqr)
    print(col)
    print(lower_bridge, upper_bridge)
    print()
```

## Dividing dependent and independent features

```python
#now we are dividing the dependent variable nad independent variable
x = data.drop(["label"], axis = 1)
y = data["label"]
print(x.shape, y.shape)
```

```python
#now lets cap the outliers and check the outliers here i am performig capping 

#In this technique, we will do the flooring (e.g., the 10th percentile) for the lower values and 
#capping (e.g., the 90th percentile) for the higher values.


for i in x :
    print(i)
    #lower = data[i].quantile(0.10)
    #upper = data[i].quantile(0.90)
    #print(lower)
    #print(upper)
    #print()
    
    #data[i] = np.where(data[i] < lower, lower,data[i])
    #data[i] = np.where(data[i] > upper, upper,data[i])
    #print(data[i].skew())
    
    #using capping using median method
    lower = x[i].quantile(0.50) 
    upper = x[i].quantile(0.95) 
    x[i] = np.where(x[i] > upper, lower, x[i])
    print(x[i].skew())
    
    #Using logtransform method to adjust skewness
    
    x[i] = x[i].map(lambda z: np.log(z) if z > 0 else 0) 
    print(x[i].skew())
    print()
    
    
```

# Applying PCA

```python
#We have already standardized the data so lets start with covarience matrix
covarience_matrix = x.cov()
covarience_matrix
```

```python
eig_values, eig_vectors = np.linalg.eig(covarience_matrix)
print("\n Number of eigrn vectors :%s\n"%len(eig_vectors))
print("\n Eigen vectors are :%s\n\n" %eig_vectors)
print("\n Eigrn value :%s" %eig_values)
```

```python
#sorting the eigen vectors in the dataset
eig_pairs = [(np.abs(eig_values[i]), eig_vectors[:,i]) for i in range(len(eig_values))]
print('Eigenvalues in descending order:')
for i in eig_pairs:
    print(i[0])
```

```python
#applying PCA function
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_data = pca.fit_transform(x)
print(x.shape)
print(x_data.shape)
```

### Undersampling:

```python
#aplying under sampling method for immbalanced dataset
from imblearn.under_sampling import RandomUnderSampler
ou = RandomUnderSampler(random_state = 42)
x_res,y_res = ou.fit_resample(x,y)
```

# Train_Test_Split

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score, RandomizedSearchCV
```

```python
#dividing datasets into testing and validations set
x_train,x_test,y_train,y_test = train_test_split(x_res, y_res, random_state = 0, test_size = 0.33)
print(x_train.shape,y_train.shape)
```

# Model Creation

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report, roc_auc_score, roc_curve
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
import warnings
warnings.filterwarnings('ignore')
```

```python
#creatinga loop for checking the crossvalscore of the data 
models = []

models.append(('Logistic_regression', LogisticRegression()))
models.append(('Guassian', GaussianNB()))
models.append(('Decisiontree', DecisionTreeClassifier()))
models.append(('Randomforest', RandomForestClassifier()))
models.append(('ADAboost', AdaBoostClassifier()))

results = []
names = []

for name, model in models:
    kfold = KFold(n_splits = 10)
    cv_results = cross_val_score(model, x_train, y_train, cv=5, scoring='accuracy')
    
    results.append(cv_results)
    names.append(name)
    
    msg = '%s: %f, %f' % (name, cv_results.mean(), cv_results.std())
    print(msg)


```

```python
#creating instances for the models
logistic_regression = LogisticRegression()
guassian = GaussianNB()
decision_tree = DecisionTreeClassifier()
random_forest = RandomForestClassifier()
adaboost = AdaBoostClassifier()


model = []
model.append(('Logistic_regression', LogisticRegression()))
model.append(('gaussian', GaussianNB()))
model.append(('decision_tree', DecisionTreeClassifier()))
model.append(('random_forest', RandomForestClassifier()))
model.append(('adaboost', AdaBoostClassifier()))


for n, m in model:
    m.fit(x_train,y_train)
    prediction = m.predict(x_test)
    print("Accuracy score of {} is {}".format(n, accuracy_score(y_test, prediction)))
    print(classification_report(y_test, prediction))
    print(confusion_matrix(y_test,prediction))
    print('========================================================================')
    
```
