```python
#importing necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
 
website_url = 'https://www.imdb.com/list/ls055592025/'
 
titles = []
years = []
imdb_ratings = []
 
source = requests.get(website_url).text
soup = BeautifulSoup(source, 'html.parser')
 
print(soup.title.text)

movie_div = soup.find_all('div', class_ = "lister-item-content")

#getting all the details from the tags
for container in movie_div:
    
    name = container.h3.a.text
    titles.append(name)
    

    year = container.h3.find('span', class_= "lister-item-year text-muted unbold").text.strip('()')

    years.append(year)
    
    imdb =container.find('span', class_ = 'ipl-rating-star__rating').text
    
    imdb_ratings.append(imdb)
    
#creating data frame for storing visuaize data into 
    
Movie_list = pd.DataFrame({
    'Name' : titles,
    'IMDB_Rating' : imdb_ratings,
    'Year' : years
})

#saving details in the form of csv file
Movie_list.to_csv("IMDB top 100 movie list.csv")
Movie_list.head()
```

```python
#importing necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd

website_url = 'https://www.imdb.com/list/ls009997493/'

names = []
imdb_rating = []
year_of_release = []

#getting the source of the file so, we can start scraping
source = requests.get(website_url).text
soup = BeautifulSoup(source, 'html.parser')

print(soup.title.text)

tags = soup.find_all('div', class_ = 'lister-item-content')

for container in tags:
    name = container.h3.a.text
    names.append(name)
    
    year = container.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.strip('()')
    year_of_release.append(year)
    
    rating = container.find('span', class_ = 'ipl-rating-star__rating').text
    imdb_rating.append(rating)
#creating Dataframe to store details

Indian_movie_list = pd.DataFrame({
    'Name' : names,
    'IMBD_rating' : imdb_rating,
    'Year' : year_of_release
})

#saing to the csv file
Indian_movie_list.to_csv("Top 100 indian movie list.csv")

Indian_movie_list.head()

```

```python
#importing necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd

website_url = 'https://bookpage.com/reviews'

book_name = []
author_name = []
genre = []
book_review = []

source = requests.get(website_url).text
soup = BeautifulSoup(source, 'html.parser')

print(soup.title.text)

book_data = soup.find_all('div', class_ = 'flex-article-content', limit=5)

#getting details from the tags

for container in book_data:
    name = container.h4.a.text
    book_name.append(name)
    
    author = container.find('p', class_ = 'sans bold').text.replace('\n','')
    author_name.append(author)
    
    book_genre = container.find('p', class_ = 'genre-links hidden-phone').text.replace('\n', '')
    genre.append(book_genre)
    
    review = container.find('p', class_="excerpt").text.replace('\n','')
    book_review.append(review)
#saving the data in csv format    
    
Book_data = pd.DataFrame({
    'Book_Name' : book_name,
    'Author_name' : author_name,
    'Genre' : genre,
    'Book-review' : book_review
})
Book_data.to_csv('Book_data.csv')
Book_data.head()
```
