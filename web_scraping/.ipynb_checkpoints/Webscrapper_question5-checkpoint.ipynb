```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"

page = requests.get(url).text
soup = BeautifulSoup(page, 'html.parser')

print(soup.title.text)

data = soup.find('table', class_ = 'table')
pos = []

posi = data.find_all('td', class_ = "rankings-block__banner--pos")


team = []
tname = data.find_all('span', class_ = 'u-hide-phablet', limit = 10)    
for i in tname:
    team.append(i.text)
    
#pos
list1 = []
pos_banner = data.find('td', class_ = 'rankings-block__banner--pos').text
list1.append(pos_banner)

list2 = []
pos_list = data.find_all('td', class_ = 'table-body__cell table-body__cell--position u-text-right', limit = 9)
for i in pos_list:
    list2.append(i.text)
    
pos = list1 + list2

#match

match = []
points = []
match_list1 = []
match_banner = data.find('td', class_ = 'rankings-block__banner--matches')
match_list1.append(match_banner.text)

basic_list =[]
match_data = data.find_all('td', class_ = 'table-body__cell u-center-text', limit = 18)
for z in match_data:
    basic_list.append(z.text)

match_list2 = basic_list[::2]


match = match_list1 + match_list2


point_data = data.find('td', class_ = 'rankings-block__banner--points')
points_list1 = []
points_list1.append(point_data.text)
points_list2 = basic_list[::-2][::-1]

points = points_list1 + points_list2

#rating
ratings = []
rating_list1 = []
rating_list2 = []
rating_data  = data.find('td', class_ = 'rankings-block__banner--rating u-text-right')
rating_list1.append(rating_data.text.strip())
rating1_data = data.find_all('td', class_ = 'table-body__cell u-text-right rating', limit = 9)
for i in rating1_data:
    rating_list2.append(i.text)
ratings = rating_list1 + rating_list2

#dataframe

Top_ODI_Teams = pd.DataFrame({
     'POS' : pos,
     'Teams' : team,
     'Matches' : match,
     'Points' : points,
     'Rating' : ratings
 })
    

Top_ODI_Teams.to_csv("Top_ODI_Teams.csv")

Top_ODI_Teams
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting"

source = requests.get(url).text

soup = BeautifulSoup(source, 'html.parser')

print(soup.title.text)

name_list1 = []
team_list1 = []
rat_list1 = []


banner_data = soup.find('tr', class_="rankings-block__banner")


name = banner_data.find('div', class_ = "rankings-block__banner--name-large").text
name_list1.append(name)

team = banner_data.find('div', class_="rankings-block__banner--nationality").text.replace('\n', '')
team_list1.append(team)

rating = soup.find('td', class_="u-text-left").text.replace('\n', '')
rat_list1.append(rating)


name_list2 = []
name_data = soup.find_all('td', class_="table-body__cell rankings-table__name name", limit = 9)
for i in name_data:
    name_list2.append(i.text.replace("\n", ''))
Name = name_list1 + name_list2



team_list2 = []
team_data = soup.find_all('span', class_="table-body__logo-text", limit = 9)
for j in team_data:
    team_list2.append(j.text)
Team = team_list1 + team_list2


rat_list2 = []
rat_data = soup.find_all('td', class_="table-body__cell rating", limit = 9)
for z in rat_data:
    rat_list2.append(z.text)
Rating = rat_list1 + rat_list2

Top_ODI_Batsman = pd.DataFrame({
    'Name' : Name,
    'Team' : Team,
    'Rating' : Rating
})

Top_ODI_Batsman.to_csv('Top_ODI_Batsman.csv')
Top_ODI_Batsman 
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling"

source = requests.get(url).text
soup = BeautifulSoup(source, 'html.parser')
print(soup.title.text)


data = soup.find('tr', class_="rankings-block__banner")

player = []
team = []
rating = []


data_player = data.find('div', class_="rankings-block__banner--name-large").text
player.append(data_player)

data_team = data.find('div', class_="rankings-block__banner--nationality").text.replace('\n', '')
team.append(data_team)

data_rating = soup.find('div', class_="rankings-block__banner--rating").text
rating.append(data_rating)

player_list = []
player_data = soup.find_all('td', class_="table-body__cell rankings-table__name name", limit = 9)
for i in player_data:
    player_list.append(i.text.replace('\n', ''))
    
Player = player + player_list
    
team_list = []
team_data = soup.find_all('span', class_="table-body__logo-text", limit = 9)
for i in team_data:
    team_list.append(i.text.replace('\n', ''))
    
Team = team + team_list
    
rating_list = []
rating_data = soup.find_all('td', class_="table-body__cell rating", limit = 9)
for i in rating_data:
    rating_list.append(i.text.replace('\n', ''))
    
Rating = rating + rating_list

Top_ODI_Bowlers = pd.DataFrame({
    'Player' : Player,
    'Team' : Team,
    'Rating' : Rating
})

Top_ODI_Bowlers.to_csv('Top_ODI_Bowlers.csv')

Top_ODI_Bowlers
```
