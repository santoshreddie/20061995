```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.amazon.in/s?bbn=1389401031&rh=n%3A1389401031%2Cp_36%3A1318506031&dc&qid=1620357122&rnid=1318502031&ref=lp_1389401031_nr_p_36_3"
class_="s-image"
source = requests.get(url).text

soup = BeautifulSoup(source, 'html.parser')

productlist = []
productName = soup.find_all('span', class_="a-size-base-plus a-color-base a-text-normal", limit=20)
for i in productName:
    productlist.append(i.text)

    
price_list = []
price_data = soup.find_all('span', class_="a-price-whole", limit=20)
for i in price_data:
    price_list.append(i.text)

    
ImageUrl = []
links = soup.find_all('img', class_="s-image", limit=20)
for link in links:
    ImageUrl.append(link.get('src'))

    
average_rating = []
avgrating = soup.find_all('span', class_="a-icon-alt", limit=20)
for rating in avgrating:
    average_rating.append(rating.text)
    
Product_list =pd.DataFrame({
    'Name' : productlist,
    'Price_In_Rupees' : price_list,
    'ImageUrl' : ImageUrl,
    'Rating' : average_rating
})



Product_list.to_csv("Product_list.csv")
Product_list




```
